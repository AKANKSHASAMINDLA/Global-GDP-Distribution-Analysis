```python
import pandas as pd

# Load the dataset
df = pd.read_csv("GDP - HW4 (1).csv")
```

```python
df
```

```python
# Define year columns (from 2000 to 2010)
year_columns = df.columns[2:]  # skipping 'Country' and 'Country Code'

# Step 1: Remove countries where all GDP values are missing
df = df.dropna(subset=year_columns, how='all')
```

```python
df.shape
```

```python
df
```

```python
# Step 2: Fill missing GDP values with the mean GDP for that country
def fill_missing_with_country_mean(row):
    gdp_values = row[year_columns]
    mean_value = gdp_values.mean(skipna=True)
    return gdp_values.fillna(mean_value)

df[year_columns] = df.apply(fill_missing_with_country_mean, axis=1)
```
